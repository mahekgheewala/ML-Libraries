Seaborn: Data Visualization Library.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built-in Datasets

```python
tip = sns.load_dataset('tips')
```

```python
tip.head()
```

```python
# Setting theme for the plots
sns.set_theme()
```

```python
# Visualize the Data (plotting)
# This function creates a relational plot, typically a scatter plot or line plot, showing relationships between two variables.

sns.relplot(data=tip, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size' )
```

```python
# Load the iris Dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Making a Scatter Plot. Predict to what species the iris flower belong(there are 3 species)

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
# here, hue = 'species' means plot will be differentiated based on species
```

```python
# House Price Dataset
from sklearn.datasets import load_diabetes
```

```python
diabete = load_diabetes()
```

```python
print(diabete)
```

```python
# Storing it in Pandas DataFrame.
diabetes = pd.DataFrame(diabete.data, columns=diabete.frame)
```

```python
diabetes.head()
```

```python
diabetes['Price'] = diabete.target
```

Distribution Plot

```python
sns.distplot(diabetes['Price'])
```

Correlation;

1)Positive correlation, 2) Negative correlation

```python
correlation = diabetes.corr()
```

Heat Map

```python
# Firstly we would define our figure size
plt.figure(figsize=(10,10))
# Now we want to make heatmap for these correlation values
sns.heatmap(correlation, cbar=True, square= True, fmt = '.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
