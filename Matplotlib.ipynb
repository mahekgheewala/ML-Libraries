Matplotlib: used for making plots.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
# y would find all the sin value of x
#  np.sin() and np.cos() are vectorized and can handle arrays as input, performing the operations element-wise on the entire array, which is crucial for numerical computations.
# In short, np is very essential for numerical operations.
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

Plotting the Data.

1) If we plot x and y, we get sin curve.

2) If we plot x and z, we get cos curve.

```python
plt.plot(x,y)
plt.show()
# By default it gives a line plot
```

```python
plt.plot(x,z)
plt.show()
```

```python
# Adding title: x-axis and y-axis.
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("sine value")
plt.title("graph")
plt.show()
```

Parabola

```python
x=np.linspace(-10, 10, 20)
# The below is the formula for parabola.
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+')
plt.show()
```

```python
# Instead of using two variables, we can plot using one variable as well.
x = np.linspace(-10, 10, 20)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python
# The below line would create an empty plot
fig = plt.figure()
# 0,0---> co-ordinates(origin), 1,1---> height and width
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'Hindi']
people = [100, 50, 45, 56, 44]
ax.bar(languages, people)
plt.xlabel("Languages")
plt.ylabel("No of people")
plt.show()
```

Pie Chart

```python
# Create a figure
fig1 = plt.figure()
# Add axes
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'Hindi']
people = [100, 50, 45, 56, 44]
# autocpt means how much decimal value we want, 1.1f means 1 decimal value
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1,])
ax.scatter(x,y, color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3-D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
# Generates 100 random numbers in the range [0, 1] using np.random.random(100).
# Multiplies these values by 20 to scale the numbers to the range [0, 20].
# This array, z, represents the z-coordinate values of the data points.
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
# c=z: This means the color of each point in the plot is decided by the value of z.
# If a point has a small z value, it will get a light blue color.
# If a point has a large z value, it will get a dark blue color.
plt.show()
```
